```python
import pandas as pd
from tabulate import tabulate

# ---------------- LOAD DATA ----------------
car_dataset = pd.read_csv("CARS_1.csv")

# Convert numeric columns correctly
car_dataset["starting_price"] = car_dataset["starting_price"].astype(int)
car_dataset["ending_price"] = car_dataset["ending_price"].astype(int)
car_dataset["rating"] = car_dataset["rating"].astype(float)

# Select required columns
car_dataset_detail = car_dataset[['Sr No.','car_name','reviews_count','fuel_type',
                                  'transmission_type','rating','starting_price','ending_price']]


def rec_rating():

    a = float(input("Enter minimum rating you want:\n"))
    b = input("Fuel type:\n")
    c = int(input("Min price:\n"))
    d = int(input("Max price:\n"))
    e = input("Transmission:\n")

    # Apply filters
    df = car_dataset_detail[
        (car_dataset_detail['rating'] >= a) &
        (car_dataset_detail['fuel_type'].str.lower() == b.lower()) &
        (car_dataset_detail['starting_price'] >= c) &
        (car_dataset_detail['ending_price'] <= d) &
        (car_dataset_detail['transmission_type'].str.lower() == e.lower())
    ]

    print("\n------------------------------------------")
    print(" ðŸ” YOUR RECOMMENDED CARS")
    print("------------------------------------------\n")

    if df.empty:
        print("âŒ No exact match found.")
    else:
        print(tabulate(df, headers="keys", tablefmt="fancy_grid", showindex=False))

    print("\n------------------------------------------")
    print("              THANK YOU ðŸ˜Š")
    print("------------------------------------------\n")


# Run
rec_rating()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import os

# replace with the uploaded filename: CARS_1.csv or CARS_1.xlsx
file_path = (r"CARS_1.csv")  # or "/mnt/data/CARS_1.xlsx"

# load
if file_path.endswith(".csv"):
    df = pd.read_csv(file_path)
else:
    df = pd.read_excel(file_path)

# coerce types
df["starting_price"] = pd.to_numeric(df.get("starting_price"), errors="coerce")
df["ending_price"] = pd.to_numeric(df.get("ending_price"), errors="coerce")
df["rating"] = pd.to_numeric(df.get("rating"), errors="coerce")
for col in ["fuel_type", "transmission_type", "car_name"]:
    if col in df.columns:
        df[col] = df[col].astype(str).str.strip()

out_dir = "/mnt/data/car_plots"
os.makedirs(out_dir, exist_ok=True)

# 1) Rating distribution (bar)
plt.figure(); df["rating"].dropna().value_counts().sort_index().plot(kind="bar")
plt.title("Rating Distribution"); plt.xlabel("Rating"); plt.ylabel("Count"); plt.tight_layout()
plt.savefig(out_dir + "/rating_distribution.png"); plt.show(); plt.close()

# 2) Price comparison scatter
plt.figure(); plt.scatter(df["starting_price"], df["ending_price"])
plt.title("Starting vs Ending Price"); plt.xlabel("Starting Price"); plt.ylabel("Ending Price")
plt.tight_layout(); plt.savefig(out_dir + "/price_scatter.png"); plt.show(); plt.close()

# 3) Fuel type count (bar)
plt.figure(); df["fuel_type"].dropna().value_counts().plot(kind="bar")
plt.title("Fuel Type Count"); plt.xlabel("Fuel Type"); plt.ylabel("Count")
plt.tight_layout(); plt.savefig(out_dir + "/fuel_count.png"); plt.show(); plt.close()

# 4) Transmission type count (bar)
plt.figure(); df["transmission_type"].dropna().value_counts().plot(kind="bar")
plt.title("Transmission Type Count"); plt.xlabel("Transmission Type"); plt.ylabel("Count")
plt.tight_layout(); plt.savefig(out_dir + "/trans_count.png"); plt.show(); plt.close()

# 5) Top 10 rated cars
top = df.dropna(subset=["rating"]).sort_values("rating", ascending=False).head(10)
plt.figure(); plt.barh(top["car_name"], top["rating"]); plt.gca().invert_yaxis()
plt.title("Top 10 Rated Cars"); plt.xlabel("Rating"); plt.tight_layout()
plt.savefig(out_dir + "/top10_rated.png"); plt.show(); plt.close()

# 6) Starting price histogram
plt.figure(); df["starting_price"].dropna().plot(kind="hist", bins=20)
plt.title("Starting Price Distribution"); plt.xlabel("Starting Price")
plt.tight_layout(); plt.savefig(out_dir + "/starting_price_hist.png"); plt.show(); plt.close()
```
